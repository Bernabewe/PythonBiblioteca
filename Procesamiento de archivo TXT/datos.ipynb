```python
# Cargamos Numpy ya que cuenta con funciones que nos va a servir para nuestro preprocesamiento de datos
import numpy as np
# Reemplazamos los valores "Nulo" por NaN para que Pandas los reconozca como valores faltantes
df.replace("Nulo", np.nan, inplace=True)
```

```python
# Mostramos el numero de nulos (ahora ya existen)
df.isnull().sum()
```

```python
# Convertimos las columnas a tipo numerico para poder trabajar con ellas
# el metodo to_numeric no fue visto en clase pero es necesario para poder trabajar con los datos
df["PRECIP"] = pd.to_numeric(df["PRECIP"])
df["EVAP"] = pd.to_numeric(df["EVAP"])
df["TMIN"] = pd.to_numeric(df["TMIN"])
df["TMAX"] = pd.to_numeric(df["TMAX"])
```

```python
# En este punto los datos ya estan listos para trabajar en ellos,
# Son tipo numericos y contiene nulos reales
df.dtypes
```

```python
# Mostramos las descripciones estadisticas una vez procesados los datos
# Como ya estan seteados los datos a tipo numerico con nulos, ahora si nos aparece la informacion completa
# del metodo describe()
df.describe()
```

```python
# "2.- Contar total de registros faltantes por fecha"

# Obtenemos todas las filas que vamos a trabajar
total_filas = df.shape[0]

# Por medio de un for interamos sobre cada fila para contar el numero de nulos que tienen por fila
for i in range(total_filas):
    fila = df.iloc[i]
    
    total_columnas = fila.shape[0]                      # Ya sabemos que es 4 pero lo correcto es calcularlo
    elementos_validos = fila.dropna().shape[0]          # Contamos solo los elementos no nulos
    faltantes = total_columnas - elementos_validos      # Calculamos el numero de nulos
    
    # Mostramos solo si hay nulos para no saturar la pantalla
    if faltantes > 0:
        print(f"{df.index[i]} | Registros faltantes: {faltantes}")
```

```python
# "1.- Eliminar rellenar nulos con la media de la columna"

# Imputamos los valores faltantes con la media de cada columna como se solicita en la asigancion
df["PRECIP"] = df['PRECIP'].fillna(df['PRECIP'].mean())
df["EVAP"] = df['EVAP'].fillna(df['EVAP'].mean())
df["TMAX"] = df['TMAX'].fillna(df['TMAX'].mean())
df["TMIN"] = df['TMIN'].fillna(df['TMIN'].mean())
```

```python
# Mostramos un resumen estadistico una vez imputados los valores faltantes
# Notamos pequeñas variaciones respecto al anterior
df.describe()
```
